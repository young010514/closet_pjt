```python
import pandas as pd
import json

df = pd.read_csv(
    "국토교통부_전국 법정동_20260609.csv",
    dtype={"법정동코드": str}
)

df = df.rename(columns={
    "시도명": "sido",
    "시군구명": "sigungu",
    "읍면동명": "dong",
    "법정동코드": "region_code",
})

df = df[["region_code", "sido", "sigungu", "dong"]]

df = df.fillna("")

# 앞뒤 공백 제거
df["sido"] = df["sido"].str.strip()
df["sigungu"] = df["sigungu"].str.strip()
df["dong"] = df["dong"].str.strip()
df["region_code"] = df["region_code"].str.strip()

# sido + sigungu + dong 기준 중복 제거
df = df.drop_duplicates(subset=["sido", "sigungu", "dong"], keep="first")

now = "2026-06-09T00:00:00+09:00"

fixture_data = []

for idx, row in df.reset_index(drop=True).iterrows():
    fixture_data.append({
        "model": "regions.region",
        "pk": idx + 1,
        "fields": {
            "region_code": row["region_code"],
            "sido": row["sido"],
            "sigungu": row["sigungu"],
            "dong": row["dong"],
            "is_active": True,
            "created_at": now,
            "updated_at": now,
        }
    })

with open("region_fixture_20260609.json", "w", encoding="utf-8") as f:
    json.dump(fixture_data, f, ensure_ascii=False, indent=4)
```
